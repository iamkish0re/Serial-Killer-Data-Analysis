# The Serial Killers

```python
import pandas as pd  # library for data analysis
import numpy as np
import requests  # library to handle requests
from bs4 import BeautifulSoup  # library to parse HTML documents
```

```python
# get the response in the form of html
wikiurl = "https://en.wikipedia.org/wiki/List_of_serial_killers_by_number_of_victims"
response = requests.get(wikiurl)
print(response.status_code)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
serialkillertable=soup.find_all('table',{'class':"wikitable"})
```

```python
serialkillertable[1]
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None
pd.options.display.max_colwidth = None
```

```python
df=pd.read_html(str(serialkillertable))
# convert list to dataframe
df=pd.DataFrame(df[0])
df.head()
```

```python
len(df)
```

```python
df_serial_killer_scrap = pd.DataFrame(
    columns=['Name', 'Country', 'Years active', 'Proven victims', 'Possible victims', 'Notes'])
```

```python
for i in serialkillertable:
    df = pd.read_html(str(i))
    # convert list to dataframe
    df = pd.DataFrame(df[0])
    df_serial_killer_scrap = df_serial_killer_scrap.append(df,
                                                           ignore_index=True)
```

```python
len(df_serial_killer_scrap)
```

```python
df_serial_killer_scrap.drop(columns = ['Unnamed: 6'], inplace = True)
```

```python
df_serial_killer_scrap
```

```python
df_serial_killer_scrap = df_serial_killer_scrap.replace('\[\d*\]', '', regex=True)
# df_serial_killer_scrap = df_serial_killer_scrap.replace(u'\xa0', u'', regex=False)
```

```python
df_serial_killer_scrap.describe()
```

```python
df_serial_killer_scrap.isna().sum()
```

```python
# Clean '\xa0' char from Country column
# df_serial_killer_scrap['Country'] = df_serial_killer_scrap['Country'].str.split().str.join(' ')
# df_serial_killer_scrap['Years active'] = df_serial_killer_scrap['Years active'].str.split().str.join(' ')
```

```python
df_serial_killer_scrap['Years active'].value_counts(dropna=False)
```

```python
years_noise_values = [
    ' and earlier', ' to present', ' to 23 July 1983', 's', '?', 'c.', 'late ', ', 1967', ','
]

for i in years_noise_values:
    df_serial_killer_scrap['Years active'] = df_serial_killer_scrap[
        'Years active'].str.replace(i, '', regex=False)

df_serial_killer_scrap['Years active'] = df_serial_killer_scrap[
    'Years active'].str.replace('30 June 1983', '1983', regex=False)
df_serial_killer_scrap['Years active'] = df_serial_killer_scrap[
    'Years active'].str.replace('-', 'to', regex=False)

df_serial_killer_scrap['Years active'] = df_serial_killer_scrap['Years active'].str.split().str.join(' ')
```

```python
df_serial_killer_scrap['Years active'].value_counts(dropna=False)
```

```python
df_serial_killer_scrap['Year From'] = np.nan
df_serial_killer_scrap['Year To'] = np.nan
```

```python
df_serial_killer_scrap.head()
```

```python
df_serial_killer_scrap[['Year From','Year To']] = df_serial_killer_scrap['Years active'].str.split('to',expand=True)
```

```python
df_serial_killer_scrap.sample(6)
```

```python
df_serial_killer_scrap.drop('Years active', axis=1, inplace=True)
df_serial_killer_scrap['Year To'].isna().sum()
```

```python
df_serial_killer_scrap['Year To'].fillna(df_serial_killer_scrap['Year From'],inplace=True)
df_serial_killer_scrap['Year To'].isna().sum()
```

```python
df_serial_killer_scrap['Name'].value_counts()
```

```python
df_serial_killer_scrap['First Name'] = np.nan
df_serial_killer_scrap['Last Name'] = np.nan

# Return a list of the words in the string, using sep as the delimiter string. 2nd param (1 in this case) tell that at most 1 split is done (thus, the list will have at most 2 elements).
df_serial_killer_scrap[['First Name','Last Name']] = df_serial_killer_scrap['Name'].str.split(' ', 1, expand=True)
```

```python
df_serial_killer_scrap['Country'].value_counts()
```

```python
country_noise_values = ['(suspected)', '(alleged)', '(possibly)', '(claimed)']

for i in country_noise_values:
    df_serial_killer_scrap['Country'] = df_serial_killer_scrap[
        'Country'].str.replace(i, '', regex=False)

df_serial_killer_scrap['Country'] = df_serial_killer_scrap[
    'Country'].str.replace('Soviet Union', 'Russia', regex=False)

df_serial_killer_scrap['Country'] = df_serial_killer_scrap[
    'Country'].str.replace('West Germany', 'Germany', regex=False)

df_serial_killer_scrap['Country'] = df_serial_killer_scrap[
    'Country'].str.replace('East Germany', 'Germany', regex=False)
df_serial_killer_scrap['Country'] = df_serial_killer_scrap[
    'Country'].str.replace('Allied-occupied Germany', 'Germany', regex=False)

df_serial_killer_scrap['Country'] = df_serial_killer_scrap[
    'Country'].str.replace('Portuguese Angola', 'Angola', regex=False)
df_serial_killer_scrap['Country'] = df_serial_killer_scrap[
    'Country'].str.replace('Ottoman Empire', 'Turkey', regex=False)
df_serial_killer_scrap['Country'] = df_serial_killer_scrap[
    'Country'].str.replace('Kingdom of Romania', 'Romania', regex=False)

df_serial_killer_scrap['Country'] = df_serial_killer_scrap[
    'Country'].str.replace('Ruanda-Urundi', 'East Africa', regex=False)
```

```python
df_serial_killer_scrap['Country'].value_counts()
```

```python
df_serial_killer_scrap['Country'] = df_serial_killer_scrap[
    'Country'].str.split('\xa0', expand=True)[0]
```

```python
df_serial_killer_scrap['Proven victims'].value_counts()
```

```python
prov_vict_noise_values = ['+', '[d]', '[c]', '[note 1]', '[b]']

for i in prov_vict_noise_values:
    df_serial_killer_scrap['Proven victims'] = df_serial_killer_scrap[
        'Proven victims'].str.replace(i, '', regex=False)
```

```python
df_serial_killer_scrap['Proven victims'] = df_serial_killer_scrap[
    'Proven victims'].str.split('-', expand=True).str.get(-1)
```

```python
temp = df_serial_killer_scrap[
    'Proven victims'].str.split('–', expand=True).str[-1]
```
